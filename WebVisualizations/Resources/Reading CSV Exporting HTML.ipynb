```python
# Dependencies
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Store filepath in a variable
file_path = "CitiesWeather.csv"
```

```python
# Read  Data file with the pandas library
citiesweatherDF = pd.read_csv(file_path, encoding="ISO-8859-1")
citiesweatherDF
```

```python
citiesweatherDF.to_html("../assets/citiesweather.html", justify='left')
```

```python
#Temperature plot
xlat = np.arange(-80, 100+1, 20)
plt.scatter(citiesweatherDF["Latitude"], citiesweatherDF["Temperature"], s=50, alpha=0.65)
plt.xticks(xlat)

#Set background colour
ax = plt.gca()
ax.set_facecolor('#e4e4e4')

#Set grid lines behind plots
ax.set_axisbelow(True)

#Remove border
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)


#set grid colour
plt.grid(color='white', linestyle='--', linewidth=0.75)

plt.title("City Latitude vs. Max Temperature (14/01/2021)")
plt.xlabel('Latitude')  
plt.ylabel('Max Temperature (c)')
plt.savefig('../assets/Lat_vs_Temp.png')
```
